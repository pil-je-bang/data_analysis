```python
import folium
import pandas as pd

```

```python
data = pd.read_csv("택시승차대 현황.csv", encoding='cp949')

m = folium.Map(location=[37.547631, 126.942463], zoom_start=15)
for index, rows in data.iterrows():
    folium.Marker([rows['위도'], rows['경도']], popup=rows['위치명'], tooltip=rows['도로명 주소'], icon=folium.Icon(color='red', icon='fa-solid fa-taxi', prefix='fa-solid')).add_to(m)
m

```
