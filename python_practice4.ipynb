```python
word = input()


wb = openpyxl.load_workbook('merchandise.xlsx')
ws = wb.active

for i in range(1, 5):
  response = requests.get(f"https://browse.gmarket.co.kr/search?keyword={word}&k=30&p={i}")
  html = response.text
  soup = BeautifulSoup(html, 'html.parser')
  products = soup.select('.box__information-major')
  for product in products:
    merchandise_tag = product.select_one('.box__item-title')
    merchandise = merchandise_tag.select_one('.text__item')['title']
    print(merchandise)

    merchandise_link = product.select_one('.link__item')['href']
    print(merchandise_link)

    merchandise_price = product.select_one('strong').text
    print(merchandise_price)

    ws.append([merchandise, merchandise_price, merchandise_link])
wb.save('merchandise.xlsx')




```

```python
from selenium.webdriver.common.keys import Keys
wb = openpyxl.load_workbook('라면 가격 비교.xlsx')
ws = wb.active

current_height = driver.execute_script("return window.scrollY")

while True:
    driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
    sleep(1)
    
    scrolled_height = driver.execute_script("return window.scrollY")
    
    if (current_height == scrolled_height):
        break
    
    current_height = scrolled_height
    
products = driver.find_elements(By.CLASS_NAME, "adProduct_item__1zC9h")

for product in products:
    name = product.find_element(By.CLASS_NAME, "adProduct_link__NYTV9.linkAnchor").text
    print(name)

    merchandise_link = product.find_element(By.CLASS_NAME, "adProduct_link__NYTV9.linkAnchor").get_attribute('href')
    print(merchandise_link)

    merchandise_price = product.find_element(By.CLASS_NAME, "price_num__S2p_v").text
    print(merchandise_price)

    ws.append([name, merchandise_price, merchandise_link])

wb.save('라면 가격 비교.xlsx')

```
